```python
import pandas as pd 
import seaborn as sns 

pokemon = pd.read_csv("../input/pokemon/Pokemon.csv", index_col=0)
pokemon.head()
```

Graph showing number of pokemons according to their Generation

```python
sns.countplot(pokemon['Generation'],palette='PuBuGn')
```

Graph showing attack and defense rate of pokemons on HEX graph with maroon color

```python
sns.jointplot(x='Attack',y='Defense',data=pokemon,kind='hex',color='maroon')
```

Preparing data for just Grass type pokemons

```python
grassType=pokemon[pokemon['Type 1']=='Grass']
grassType.head()
```

Just showing the Attack and Defense rate of grass pokemons in scatterplot graph, with green color.

```python
sns.jointplot(x='Attack',y='Defense',data=grassType,color='g')
```

Faceted graphs which showing attacks of all types of pokemons.

```python
p=sns.FacetGrid(pokemon,col='Type 1',col_wrap=6)
p.map(sns.countplot,'Attack')
```
